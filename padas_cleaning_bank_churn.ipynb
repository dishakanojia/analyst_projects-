```python
df["exit_date"] = pd.to_datetime(df["exit_date"])            # NaT = still active
df["signup_date"] = pd.to_datetime(df["signup_date"])
df["last_complaint_date"] = pd.to_datetime(df["last_complaint_date"])
 
df["recent_avg_txns"] = df["recent_avg_txns"].fillna(0)      # exited: no recent txns
df["activity_decline_pct"] = df["activity_decline_pct"].fillna(100)  # full stop = 100% decline
df["has_complained"] = df["complaint_count"].gt(0).astype(int) 
df.isnull().sum()      # cleaner flag
```

```python
#Duplicates & Validity
df["customer_id"].duplicated().sum()
df.duplicated().sum()
((df.age < 18) | (df.age > 100)).sum()
(df.balance < 0).sum()
bad_exit = ((df.exited == 1) & df.exit_date.isna()).sum() + \
           ((df.exited == 0) & df.exit_date.notna()).sum()
#Logical consistency check — my favorite. & means AND. Two impossible
#combinations: (churned AND no exit date) + (active AND has exit date).
#Both counts should be 0. The \ just continues the line.
#This checks whether columns agree with each other — a step most
#beginners skip and interviewers love.
```

```python
 #Standardisation & New Features
df["gender"] = df["gender"].str.strip().str.title()
df["age_band"] = pd.cut(df.age, bins=[17, 30, 40, 50, 60, 200],
                        labels=["18-30", "31-40", "41-50", "51-60", "60+"])
df["balance_lakh"] = (df.balance / 1e5).round(2)
df["is_zero_balance"] = (df.balance == 0).astype(int)
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace('  ','_')
```
